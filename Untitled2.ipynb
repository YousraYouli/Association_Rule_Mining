```python
import pandas as pd
dataframe= pd.read_csv("data/Groceries_dataset.csv")
dataframe
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

# 2. Prétraitement: Grouper les articles par chaque transaction
transactions = dataframe.groupby('Member_number')['itemDescription'].apply(list)
transactions
```

```python
# 3. Convertir les transactions au format One-Hot Encoding
encoder = TransactionEncoder()
pred = encoder.fit_transform(transactions)
df = pd.DataFrame(pred,columns = encoder.columns_)
df
```

```python
# df.to_csv("cleaned_basket_data.csv" ,index =False)
```

```python
# Find frequent itemsets : support
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)
print("frequent_itemsets :" )
frequent_itemsets
```

```python
# Generate association rules
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1 ,num_itemsets=2)
# rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=1 ,num_itemsets=2)
#.head(50)
rules
```

```python
# Afficher des statistiques supplémentaires sur les règles générées
print("### Statistiques sur les Règles d'Association")
print(f"Nombre total de règles générées: {len(rules)}")
print(f"Règles avec un lift supérieur à 1 (indiquant des associations intéressantes):")
rules[rules['lift'] > 1].head(50)
```

```python
def predict(new_transaction):
    recommendations = set()  # Use a set to eliminate duplicates
    for index, row in rules.iterrows():
        antecedents = row['antecedents']
        consequents = row['consequents']
        if all(item in new_transaction for item in antecedents):
            recommendations.add(tuple(consequents))  # Add tuples to the set
    return list(recommendations)
new_transaction = ['Milk', 'Yogurt']
recommendations = predict(new_transaction)
recommendations
```

```python
import pickle
results= list(rules)  

filename = 'apyiori_model.ob' 
pickle.dump(results, open(filename, 'wb')) 

loaded_model = pickle.load(open(filename, 'rb')) 
print(loaded_model)
```
